## 5.Perform meaningful visualizations using the dataset. Bring at least 3 visualizations

### Scatter plot to compare between different species based on their sepal and petal measurements

```python
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.title('Comparison between species based on Sepal Length and Sepal Width')
sns.scatterplot(x=iris_data['SL'],y=iris_data['SW'],hue=iris_data['Classification'])

plt.subplot(2,1,2)
plt.title('Comparison between species based on Petal Length and Petal Width')
sns.scatterplot(x=iris_data['PL'],y=iris_data['PW'],hue=iris_data['Classification'])
```

### Boxplot to analyse the data to find the outliers and variation in the data 

```python
length_width = iris_data[['PL','PW','SL','SW']] 
length_width.boxplot()
plt.xlabel('Flower measurements')
plt.ylabel('values')
plt.title("Iris dataset analysis")
plt.show()
```

### Heatmap to find the correlation of features in data

```python
sns.heatmap(iris_data.corr(), annot=True)
plt.show()
```

### Pairplot to visualise pairwise relationship of every column in the dataset

```python
sns.pairplot(iris_data)
```
